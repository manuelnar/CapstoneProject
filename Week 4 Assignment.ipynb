# Capstone Project - The Battle of Neighborhoods

## Week 1 - Part 1: Business Problem

### If we consider the Business of Restaurants, one of the biggest issues they have to deal with is deciding where to open their local. The idea of this project is to realize a recommendation app for new Restuarant owners. This will be made with the analysis of geographical Data via Foursquare to recommend the best possibilities in the different neighborhoods in the city of New York.



### The idea of this app is to give all the information about the different restaurants in every neighborhood so the new owner can make a informed decision considering which type of restaurants are doing well, which type of cousine is not present in the area of interest, which places are the ones with most quantities of restaurants. 

### All this information will be used by the new owner to make an informed decision and be closest to achieve a profitable business.
